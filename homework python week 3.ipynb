```python
df_num = df.select_dtypes(include = ['float64', 'int64'])
df_num.head()
```

```python
df_num.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8);
```

```python
df_num_corr = df_num.corr()['Fare'][:-1] # -1 because the latest row is fare
golden_features_list = df_num_corr[abs(df_num_corr) > 0.5].sort_values(ascending=False)
print("There is {} strongly correlated values with Fare:\n{}".format(len(golden_features_list), golden_features_list))
print('it will be interesting to understand the correlation between the class of the passanger, how much do they paid and if they survived or not')
```

```python
# using apply function to create a new column 
df['doubledFare'] = df.apply(lambda row: (row.Fare * 2), axis = 1) 
```

```python
corr = df_num.drop('Fare', axis=1).corr() 
plt.figure(figsize=(12, 10))

sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)], 
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True);
```

```python
quantitative_features_list = ['PassengerId', 'Survived','Pclass','Age','SibSp','Parch','Fare']
df_quantitative_values = df[quantitative_features_list]
df_quantitative_values.head()
```

```python
features_to_analyse = [x for x in quantitative_features_list if x in golden_features_list]
features_to_analyse.append('Age')
features_to_analyse
```

```python
fig, ax = plt.subplots(round(len(features_to_analyse) / 3), 3, figsize = (18, 12))

for i, ax in enumerate(fig.axes):
    if i < len(features_to_analyse) - 1:
        sns.regplot(x=features_to_analyse[i],y='Age', data=df[features_to_analyse], ax=ax)

print ('The older people was in first class,while the younger traveled in third class')
```
